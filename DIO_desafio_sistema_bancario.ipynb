# **DIO.ME - DESAFIO: SISTEMA BANCÁRIO**

```python
# ====================================
# DIO.ME - DESAFIO: SISTEMA BANCÁRIO
# AUTOR: EDUARDO BRESSAN
# ====================================

# Variáveis e constantes
saldo = 0
limite_saque_valor = 500
quantidade_saques = 0
LIMITE_SAQUES = 3
extrato = ""

# Menu principal do banco
def exibir_menu():
  print(" MENU ".center(32,"="))
  print("[1] Depositar")
  print("[2] Sacar")
  print("[3] Extrato")
  print("[4] Sair")
  return input("Escolha uma das opcoes acima: ")

# Laço principal do programa
while True:

  opcao = exibir_menu()

  # Menu: Depósito
  if opcao == "1":

    valor_deposito = float(input("Informe o valor do depósito: "))

    if valor_deposito > 0:
      saldo += valor_deposito
      extrato += f"Depósito:".ljust(9," ") +  f"R$ {valor_deposito:.2f}\n".rjust(19," ")
      print("Depósito realizado com sucesso!\n")

    else:
      print("ERRO! Valor informado inválido.")

  # Menu: Saque
  elif opcao == "2":

    valor_saque = float(input("Informe o valor do saque: "))

    erro_saldo = valor_saque > saldo
    erro_limite = valor_saque > limite_saque_valor
    erro_qtd_saques = quantidade_saques >= LIMITE_SAQUES

    if erro_qtd_saques:
      print("\n")
      print("ERRO! Limite da quantidade de saques atingido.")
      print("\n")

    elif erro_saldo:
      print("\n")
      print("ERRO! Saldo insuficiente.")
      print("\n")

    elif erro_limite:
      print("\n")
      print("ERRO! Limite do valor de saque excedido.")
      print("\n")

    elif valor_saque > 0:
      saldo -= valor_saque
      extrato += f"Saque:".ljust(9," ") + f"R$ {valor_saque:.2f}\n".rjust(19," ")
      quantidade_saques += 1
      print("\n")
      print("Saque realizado com sucesso!")
      print("\n")

    else:
      print("\n")
      print("ERRO! Valor inválido.")
      print("\n")


  # Menu: Extrato
  elif opcao == "3":
    print("")
    print(" Extrato ".center(28,"="))
    print("Não foram realizados movimentações." if not extrato else extrato)
    print(f"Saldo:".ljust(9," ") + f"R$ {saldo:.2f}".rjust(19," "))
    print("============================\n")

  # Menu: Sair
  elif opcao == "4":
    print("\n")
    print("Agradecemos a utilização de nosso sistema bancário.")
    break

  else:
    print("\n")
    print("ERRO! Valor informado inválido.")
    print("\n")
```
